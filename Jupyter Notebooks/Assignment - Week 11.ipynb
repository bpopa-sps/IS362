### Parsing the response

Next I will create a method where I will input the API response, it will loop through it, and generate a data frame.


```python
## Defining the method and its input
def parse_response(response):
    ## Creating a empty data frame containing 4 columns (list name, oldest date, newest date, and frequency)
    data = {'List Name': [],  
        'Oldest Published Date': [], 
        'Newest Published Date': [],
        'Update Frequency': []}
   
    ## Assigning the input data to the results variable
    results = response['results'] 
    
    ## Using a FOR loop to passthrough the input data and append each entry to the data frame
    for result in results: 
            data['List Name'].append(result['display_name'])
            data['Oldest Published Date'].append(result['oldest_published_date'])
            data['Newest Published Date'].append(result['newest_published_date'])
            data['Update Frequency'].append(result['updated'])
            
    ## Output the data frame
    return pd.DataFrame(data) 
```
