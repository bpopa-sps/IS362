## 3. On February 12th, 2013, which New York area airport had the windiest weather?

To find out the windiest weather we have to look at the weather.csv dataset and calculate the wind speed average for each NYC airport.


```python
#Import data from csv
wdata = pd.read_csv ('weather.csv')

#Dispaly table
wdata
```

Next, I filter the data by origin and select only the information about February 12th, 2013.


```python
# Creating a new data frame with the EWR data for 2/12/2013
wd_ewr = wdata[(wdata['year']==2013) & (wdata['month']==2) & (wdata['day']==12) & (wdata['origin']=='EWR')]

# LGA Data Frame
wd_lga = wdata[(wdata['year']==2013) & (wdata['month']==2) & (wdata['day']==12) & (wdata['origin']=='LGA')]

# JFK Data Frame
wd_jfk = wdata[(wdata['year']==2013) & (wdata['month']==2) & (wdata['day']==12) & (wdata['origin']=='JFK')]
  
# Extract only the Wind Speed from each data frame
# EWR wind speed
ws_ewr = wd_ewr['wind_speed']

# LGA wind speed
ws_lga = wd_lga['wind_speed']

# JFK wind speed
ws_jfk = wd_jfk['wind_speed']
```

#### Dealing with outliners

To remove the outliners, I have decided to trim the lowest and highest quantile from the data.

The wind speed is measured every hour so there are 24 entries for 2/12/2013. Trimming the lowest and highest entry will leave the us with 22 entries.

```python
# Trimming the lowest and highest quantile for each data frame.
ewr_tr = ws_ewr[ws_ewr.between(ws_ewr.quantile(.01), ws_ewr.quantile(.99))]
lga_tr = ws_lga[ws_lga.between(ws_lga.quantile(.01), ws_lga.quantile(.99))]
jfk_tr = ws_jfk[ws_jfk.between(ws_jfk.quantile(.01), ws_jfk.quantile(.99))]
```

To find out which NYC airport was the windies on 2/12/2013 I will calculate and compare the mean of the recorded wind speed.

```python
print("Wind Speed Mean at EWR:",ewr_tr.mean())
print("Wind Speed Mean at LGA:",lga_tr.mean())
print("Wind Speed Mean at JFK:",jfk_tr.mean())
```

### Q3 Answer: LGA

The LGA airport was the windiest airport in NYC on 2/12/2013.
