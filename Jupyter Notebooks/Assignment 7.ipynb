```python
#import pandas
import pandas as pd

#Import data from csv
data = pd.read_csv ('MovieSurvey.csv')

# display the dataframe
data
```

Calculating the average of each row and storing the result in a new colulmn.

Since there are nul values in the data the calculation needs to skip na value and look only at the numeric values.

```python

# Creating a new "Average Rating" column and calculating the mean of each row
data['Average Rating'] = data.mean(axis=1, numeric_only=True, skipna=True)

# display the dataframe
data
```

Next, we calculate the average of each column and store it on new row called Avg

```python
# Adding a new row and calculating the mean of each column
data.loc['Average'] = data.mean(axis=0, numeric_only=True, skipna=True)

# display the dataframe
data
```

### Data Normalization

To normalize the data we use the min/max formula which scales each velue in the range of 0 to 1.

x = (x-x_min)/(x_max-x_min)

Again we are only looking at the numberic values of the data and skipping nul.


```python
#Resetting the data by reimporting it from csv
nd = pd.read_csv ('MovieSurvey.csv')

# Skipping the first column 
ndata=nd.iloc[:,1:]

#Normalizing all data in the datagframe while skipping the 1st row
nd.iloc[:,1:]=(ndata-ndata.min(numeric_only=True, skipna=True))/(ndata.max(numeric_only=True, skipna=True)-ndata.min())

nd
```

Next we calculate the averages of each row and colulmn.

```python
# Calculating the averages of the normalized data
nd['Average Rating'] = nd.mean(axis=1, numeric_only=True, skipna=True)
nd.loc['Average'] = nd.mean(axis=0, numeric_only=True, skipna=True)

nd
```

Data normalization brings all data value in to a standarized range making it easier to analyze. In the case of the min-man normalizaiton the data is brought to a value between 0 and 1.

Normalization scales the value down and improves the performace of the data set.

A disadvantage of normalization is that the data set will no longer include the real values that were collected. In this case, were we looked at movie ratings, after normalization we can no longer tell what rating each person gave to the movies.
